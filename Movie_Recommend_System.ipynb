```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  count=0
  l=[]
  for i in ast.literal_eval(obj):
    if count!=3:
      l.append(i['name'])
      count+=1
    else:
      break

  return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
movies.head()
```

```python
movies['crew'].values
```

```python
def find_dir(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if(i['job']=='Director'):
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(find_dir)
```

```python
movies['crew']
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
df=movies[['movie_id','title','tags']].copy()
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags'][0]
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
df['tags'][0]
```

```python
!pip install nltk
import nltk
nltk.download('punkt')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags']=df["tags"].apply(stem)
```

```python
df['tags'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vec=cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
vec
```

```python
from sklearn.metrics.pairwise import cosine_similarity
simi=cosine_similarity(vec)
simi
```

```python
simi.shape
```

```python
df['title']=df['title'].str.lower()
```

```python
df[df['title']=='batman begins'].index[0]
```

```python
def recommend(movie):
  movie_index=df[df['title']==movie.lower()].index[0]
  dist=simi[movie_index]
  movies_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(movies.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
